```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dir_results = "../../resnet_patch_padding_sgd/"
```

```python
df_resnet_18 = pd.read_csv(os.path.join(dir_results, "fold_1/resnet_18_deeplab_v3+/train_metrics.csv"))
df_resnet_34 = pd.read_csv(os.path.join(dir_results, "fold_3/resnet_34_deeplab_v3+/train_metrics.csv"))
df_resnet_50 = pd.read_csv(os.path.join(dir_results, "fold_5/resnet_50_deeplab_v3+/train_metrics.csv"))
df_resnet_101 = pd.read_csv(os.path.join(dir_results, "fold_4/resnet_101_deeplab_v3+/train_metrics.csv"))
```

```python
dir_results_eff_net = "../../efficientnet_v2_patch_padding_sgd/"
```

```python
df_effnet_v2_s = pd.read_csv(os.path.join(dir_results_eff_net, "fold_4/efficientnet_v2_s_deeplab_v3/train_metrics.csv"))
df_effnet_v2_m = pd.read_csv(os.path.join(dir_results_eff_net, "fold_3/efficientnet_v2_m_deeplab_v3/train_metrics.csv"))
```

## ResNet18 - DeepLab_v3+

```python
df_resnet_18.head()
```

```python
fig = plt.figure(figsize=(12, 8))
plt.plot(np.arange(1, df_resnet_18.shape[0]+1), df_resnet_18.train_loss, label="train loss")
plt.plot(np.arange(1, df_resnet_18.shape[0]+1), df_resnet_18.valid_loss, label="validation loss")
plt.title("train/validation loss vs epoch for resnet18 - deeplab_v3+ model", fontsize=20)
plt.xlabel("epoch", fontsize=20)
plt.ylabel("loss", fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
plt.grid()
fig.show()
```

```python
fig = plt.figure(figsize=(12, 8))
plt.plot(np.arange(1, df_resnet_18.shape[0]+1), df_resnet_18.valid_IOU, label="validation IoU")
plt.title("validation IoU vs epoch for resnet18 - deeplab_v3+ model", fontsize=20)
plt.xlabel("epoch", fontsize=20)
plt.ylabel("IoU score", fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
plt.grid()
fig.show()
```

```python
fig = plt.figure(figsize=(12, 8))
plt.plot(np.arange(1, df_resnet_18.shape[0]+1), df_resnet_18.valid_acc, label="validation accuracy")
plt.title("validation accuracy vs epoch for resnet18 - deeplab_v3+ model", fontsize=20)
plt.xlabel("epoch", fontsize=20)
plt.ylabel("accuracy score", fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
plt.grid()
fig.show()
```
