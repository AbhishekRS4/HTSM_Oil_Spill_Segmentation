```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_resnet_18 = pd.read_csv("../../resnet_18_deeplab_v3+_zero/train_metrics.csv")
df_resnet_34 = pd.read_csv("../../resnet_34_deeplab_v3+_zero/train_metrics.csv")
df_resnet_50 = pd.read_csv("../../resnet_50_deeplab_v3+_zero/train_metrics.csv")
df_resnet_101 = pd.read_csv("../../resnet_101_deeplab_v3+_zero/train_metrics.csv")
```

```python
df_resnet_18.head()
```

```python
fig = plt.figure(figsize=(12, 8))
plt.plot(np.arange(1, df_resnet_18.shape[0]+1), df_resnet_18.train_loss, label="train loss")
plt.plot(np.arange(1, df_resnet_18.shape[0]+1), df_resnet_18.valid_loss, label="validation loss")
plt.title("train/validation loss vs epoch for resnet18 - deeplabv3+ model", fontsize=20)
plt.xlabel("epoch", fontsize=20)
plt.ylabel("loss", fontsize=20)
plt.legend(fontsize=20)
plt.grid()
fig.show()
```

```python
fig = plt.figure(figsize=(12, 8))
plt.plot(np.arange(1, df_resnet_18.shape[0]+1), df_resnet_18.valid_acc, label="validation accuracy")
plt.title("validation accuracy vs epoch for resnet18 - deeplabv3+ model", fontsize=20)
plt.xlabel("epoch", fontsize=20)
plt.ylabel("accuracy score", fontsize=20)
plt.legend(fontsize=20)
plt.grid()
fig.show()
```
